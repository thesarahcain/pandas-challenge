```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os
```

```python
# Filepaths
school_file_path = os.path.join('Resources','schools_complete.csv')
student_file_path = os.path.join('Resources','students_complete.csv')

# Read files then convert them to a dataframe 
school_data = pd.read_csv(school_file_path)
student_data = pd.read_csv(student_file_path)

# Combine data into one dataframe
final_school_data = pd.merge(student_data, school_data, how="left",on=["school_name", "school_name"])
final_school_data.head()
```

```python
# District Summary 

# Find the total number of unique schools 

total_schools = len(final_school_data['School ID'].unique())
total_schools
```

```python
# Find the total number of students

total_students = len(final_school_data['Student ID'].unique())
total_students
```

```python
# Find the total budget 

total_budget =sum(final_school_data['budget'])
total_budget
```

```python
# Find the average math score 

avg_math_score = np.mean(student_data['math_score'])
avg_math_score
```

```python
# Find the average reading score 

avg_reading_score = np.mean(student_data['reading_score'])
avg_reading_score
```

```python
# Find the percentage of students who passed math 

numerator = sum(student_data['math_score']>=70)
denominator = len(student_data['math_score'])
percent_passing_math = (numerator / denominator) * 100 
print(percent_passing_math,"%")
```

```python
# Find the percentage of students who passed reading 

numerator = sum(student_data['reading_score']>=70)
denominator = len(student_data['reading_score'])
percent_passing_reading = (numerator / denominator) * 100 
print(percent_passing_reading, "%")
```

```python
# Find the percentage of students who passed reading and math 

numerator = sum((student_data['reading_score'] >= 70) & (student_data['math_score'] >= 70))
denominator = len(student_data['reading_score'])
percent_passing_both = (numerator / denominator) * 100 
print(percent_passing_both, "%")
```

```python
# Make a snapshot of the district's key metrics in a Data Frame 

summary_data = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "Percent Passing Math": [percent_passing_math],
    "Percent Passing Reading": [percent_passing_reading],
    "Percent Passing Both": [percent_passing_both]
})
summary_data["Total Students"] = summary_data["Total Students"].map("{:,}".format)
summary_data["Total Budget"] = summary_data["Total Budget"].map("${:,.2f}".format)

summary_data

```

```python
# School Summary 

school_name = school_data['school_name']
school_name
```

```python
# School Type 

school_type = school_data['type']
school_type
```

```python
# Total Students per school

students_per_school = final_school_data.groupby(['school_name'])['Student ID'].size()
students_per_school
```

```python
# School Budget per School 

budget_per_school = final_school_data.groupby(['school_name'])['budget'].first()
budget_per_school
```

```python
# Budget per student per school 

per_student_budget_numerator = final_school_data.groupby(['school_name'])['budget'].first()
per_student_budget_denominator = final_school_data.groupby(['school_name'])['Student ID'].size()
per_student_budget = per_student_budget_numerator / per_student_budget_denominator
per_student_budget
```

```python
# Average Math Score per School

avg_math_score_per_school = final_school_data.groupby(['school_name'])['math_score'].mean()
avg_math_score_per_school 
```

```python
# Average Reading Score per School

avg_reading_score_per_school = final_school_data.groupby(['school_name'])['reading_score'].mean()
avg_reading_score_per_school 
```

```python
# Percent of students who passed math per school
students_passing_math_school = final_school_data[(final_school_data['math_score']>=70)].groupby('school_name')["student_name"].count()
students_passing_math_school_percent = (students_passing_math_school / students_per_school) * 100
students_passing_math_school_percent
```

```python
# Percent of students who passed reading per school 
students_passing_reading_school = final_school_data[(final_school_data['reading_score']>=70)].groupby('school_name')["student_name"].count()
students_passing_reading_school_percent = (students_passing_reading_school / students_per_school) * 100
students_passing_reading_school_percent
```

```python
# Percent of students who passed reading and math per school 
passing_math_and_reading = final_school_data[(final_school_data['reading_score'] >= 70) & (final_school_data['math_score'] >= 70)]
students_passing_math_and_reading_school = passing_math_and_reading.groupby('school_name')['student_name'].count()
students_passing_math_and_reading_school_percent = (students_passing_math_and_reading_school / students_per_school) * 100
students_passing_math_and_reading_school_percent
```

```python
# Convert to data frame
per_school_summary = pd.DataFrame({"School Type": school_type,
                                   "Total Students": students_per_school,
                                   "Total School Budget": budget_per_school,
                                   "Per Student Budget": per_student_budget,
                                   "Average Math Score": avg_math_score_per_school,
                                   "Average Reading Score": avg_reading_score_per_school,
                                   "% Passing Math": students_passing_math_school_percent,
                                   "% Passing Reading": students_passing_reading_school_percent,
                                   "% Overall Passing Rate": students_passing_math_and_reading_school_percent})

# Minor data munging
per_school_summary = per_school_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget",
                                         "Average Math Score", "Average Reading Score", 
                                         "% Passing Math", "% Passing Reading", 
                                         "% Overall Passing Rate"]]
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the data frame
per_school_summary
```

```python
# Sort and show the highest performing schools

top_schools = per_school_summary.sort_values(["% Overall Passing Rate"], ascending=False)
top_schools.head(5)
```

```python
# Sort and show the lowest performing schools 

low_schools = per_school_summary.sort_values(["% Overall Passing Rate"], ascending=True)
low_schools.head(5)
```

```python
# Math Scores by Grade

# Create data series of scores by grade levels
nineth_graders = final_school_data[(final_school_data["grade"] == "9th")]
tenth_graders = final_school_data[(final_school_data["grade"] == "10th")]
eleventh_graders = final_school_data[(final_school_data["grade"] == "11th")]
twelfth_graders = final_school_data[(final_school_data["grade"] == "12th")]

# Group schools by name
nineth_graders_scores = nineth_graders.groupby(["school_name"])["math_score"].mean()
tenth_graders_scores = tenth_graders.groupby(["school_name"])["math_score"].mean()
eleventh_graders_scores = eleventh_graders.groupby(["school_name"])["math_score"].mean()
twelfth_graders_scores = twelfth_graders.groupby(["school_name"])["math_score"].mean()

# Combine into single data frame
scores_by_grade = pd.DataFrame({"9th": nineth_graders_scores, "10th": tenth_graders_scores,
                                "11th": eleventh_graders_scores, "12th": twelfth_graders_scores})

# Data Cleaning
scores_by_grade = scores_by_grade[["9th", "10th", "11th", "12th"]]
scores_by_grade.index.name = None

scores_by_grade
```

```python
# Reading Scores by Grade

# Create data series of scores by grade levels
nineth_graders = final_school_data[(final_school_data["grade"] == "9th")]
tenth_graders = final_school_data[(final_school_data["grade"] == "10th")]
eleventh_graders = final_school_data[(final_school_data["grade"] == "11th")]
twelfth_graders = final_school_data[(final_school_data["grade"] == "12th")]

# Group schools by name
nineth_graders_scores = nineth_graders.groupby(["school_name"])["reading_score"].mean()
tenth_graders_scores = tenth_graders.groupby(["school_name"])["reading_score"].mean()
eleventh_graders_scores = eleventh_graders.groupby(["school_name"])["reading_score"].mean()
twelfth_graders_scores = twelfth_graders.groupby(["school_name"])["reading_score"].mean()

# Combine into single data frame
scores_by_grade = pd.DataFrame({"9th": nineth_graders_scores, "10th": tenth_graders_scores,
                                "11th": eleventh_graders_scores, "12th": twelfth_graders_scores})

# Data Cleaning
scores_by_grade = scores_by_grade[["9th", "10th", "11th", "12th"]]
scores_by_grade.index.name = None

scores_by_grade
```

```python
# Scores by School Spending 

# Define the spending bins 

spending_bins = [0, 585, 630, 645, 680]

groups = ["Less than $585","$580-$630", "$630-645", "$645-680"]

# Categorize spending based on the bins
per_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=groups)

spending_math_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = per_school_summary.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = per_school_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = per_school_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_rate = (spending_passing_math + spending_passing_reading) / 2

# Assemble into data frame
spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})


# Minor data munging
spending_summary = spending_summary[["Average Math Score", 
                                     "Average Reading Score", 
                                     "% Passing Math", "% Passing Reading",
                                     "% Overall Passing Rate"]]

# Display results
spending_summary
```

```python
# Establish the bins 
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Categorize the spending based on the bins
per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=group_names)

# Calculate the scores based on bins
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
overall_passing_rate = (size_passing_math + size_passing_reading) / 2

# Assemble into data frame
size_summary = pd.DataFrame({"Average Math Score" : size_math_scores,
                             "Average Reading Score": size_reading_scores,
                             "% Passing Math": size_passing_math,
                             "% Passing Reading": size_passing_reading,
                             "% Overall Passing Rate": overall_passing_rate})

# Minor data munging
size_summary = size_summary[["Average Math Score", 
                             "Average Reading Score", 
                             "% Passing Math", "% Passing Reading",
                             "% Overall Passing Rate"]]

# Display results
size_summary
```

```python
# Type: Average Math Score, Average Reading Score, % Passing Math, % Passing Reading, % Overall Passing Rate

type_math_scores = per_school_summary.groupby(["School Type"])["Average Math Score"].mean()
type_reading_scores = per_school_summary.groupby(["School Type"])["Average Reading Score"].mean()
type_passing_math = per_school_summary.groupby(["School Type"])["% Passing Math"].mean()
type_passing_reading = per_school_summary.groupby(["School Type"])["% Passing Reading"].mean()
overall_passing_rate = (type_passing_math + type_passing_reading) / 2

# Assemble into data frame
type_summary = pd.DataFrame({"Average Math Score" : type_math_scores,
                             "Average Reading Score": type_reading_scores,
                             "% Passing Math": type_passing_math,
                             "% Passing Reading": type_passing_reading,
                             "% Overall Passing Rate": overall_passing_rate})

# Minor data munging
type_summary = type_summary[["Average Math Score", 
                             "Average Reading Score",
                             "% Passing Math",
                             "% Passing Reading",
                             "% Overall Passing Rate"]]

# Display results
type_summary
```
